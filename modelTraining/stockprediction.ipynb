```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as data
import yfinance as yf
```

```python
startDate = '2004-01-01'
endDate = '2024-11-23'

googl = yf.download('GOOGL', startDate, endDate)
aapl = yf.download('AAPL', startDate, endDate)
msft = yf.download('MSFT', startDate, endDate)
nvda = yf.download('NVDA', startDate, endDate)
amzn = yf.download('AMZN', startDate, endDate)
meta = yf.download('META', startDate, endDate)
nflx = yf.download('NFLX', startDate, endDate)

googl.to_csv("GOOGL.csv")
aapl.to_csv("AAPL.csv")
msft.to_csv("MSFT.csv")
nvda.to_csv("NVDA.csv")
amzn.to_csv("AMZN.csv")
meta.to_csv("META.csv")
nflx.to_csv("NFLX.csv")
```

```python
from sklearn.preprocessing import MinMaxScaler
googl_scaler = MinMaxScaler(feature_range=(0, 1))
aapl_scaler = MinMaxScaler(feature_range=(0, 1))
msft_scaler = MinMaxScaler(feature_range=(0, 1))
amzn_scaler = MinMaxScaler(feature_range=(0, 1))
nvda_scaler = MinMaxScaler(feature_range=(0, 1))
meta_scaler = MinMaxScaler(feature_range=(0, 1))
nflx_scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
train_googl_arr = googl_scaler.fit_transform(train_googl)
train_aapl_arr = aapl_scaler.fit_transform(train_aapl)
train_msft_arr = msft_scaler.fit_transform(train_msft)
train_amzn_arr = amzn_scaler.fit_transform(train_amzn)
train_nvda_arr = nvda_scaler.fit_transform(train_nvda)
train_meta_arr = meta_scaler.fit_transform(train_meta)
train_nflx_arr = nflx_scaler.fit_transform(train_nflx)
```

```python
x_googl_train = []
y_googl_train = []
for i in range(100, train_googl_arr.shape[0]):
    x_googl_train.append(train_googl_arr[i-100: i])    
    y_googl_train.append(train_googl_arr[i, 0])
x_googl_train, y_googl_train = np.array(x_googl_train), np.array(y_googl_train)

x_amzn_train = []
y_amzn_train = []
for i in range(100, train_amzn_arr.shape[0]):
    x_amzn_train.append(train_amzn_arr[i-100: i])    
    y_amzn_train.append(train_amzn_arr[i, 0])
x_amzn_train, y_amzn_train = np.array(x_amzn_train), np.array(y_amzn_train)

x_nflx_train = []
y_nflx_train = []
for i in range(100, train_nflx_arr.shape[0]):
    x_nflx_train.append(train_nflx_arr[i-100: i])    
    y_nflx_train.append(train_nflx_arr[i, 0])
x_nflx_train, y_nflx_train = np.array(x_nflx_train), np.array(y_nflx_train)

x_aapl_train = []
y_aapl_train = []
for i in range(100, train_aapl_arr.shape[0]):
    x_aapl_train.append(train_aapl_arr[i-100: i])    
    y_aapl_train.append(train_aapl_arr[i, 0])
x_aapl_train, y_aapl_train = np.array(x_aapl_train), np.array(y_aapl_train)

x_meta_train = []
y_meta_train = []
for i in range(100, train_meta_arr.shape[0]):
    x_meta_train.append(train_meta_arr[i-100: i])    
    y_meta_train.append(train_meta_arr[i, 0])
x_meta_train, y_meta_train = np.array(x_meta_train), np.array(y_meta_train)

x_msft_train = []
y_msft_train = []
for i in range(100, train_msft_arr.shape[0]):
    x_msft_train.append(train_msft_arr[i-100: i])    
    y_msft_train.append(train_msft_arr[i, 0])
x_msft_train, y_msft_train = np.array(x_msft_train), np.array(y_msft_train)

x_nvda_train = []
y_nvda_train = []
for i in range(100, train_nvda_arr.shape[0]):
    x_nvda_train.append(train_nvda_arr[i-100: i])    
    y_nvda_train.append(train_nvda_arr[i, 0])
x_nvda_train, y_nvda_train = np.array(x_nvda_train), np.array(y_nvda_train)
```

```python
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
googl_model = Sequential()

googl_model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_googl_train.shape[1], 1)))
googl_model.add(Dropout(0.2))

googl_model.add(LSTM(units=60, activation='relu', return_sequences=True))
googl_model.add(Dropout(0.3))

googl_model.add(LSTM(units=80, activation='relu', return_sequences=True))
googl_model.add(Dropout(0.4))

googl_model.add(LSTM(units=120, activation='relu'))
googl_model.add(Dropout(0.5))

googl_model.add(Dense(units=1))
#-----------------------------
aapl_model = Sequential()

aapl_model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_aapl_train.shape[1], 1)))
aapl_model.add(Dropout(0.2))

aapl_model.add(LSTM(units=60, activation='relu', return_sequences=True))
aapl_model.add(Dropout(0.3))

aapl_model.add(LSTM(units=80, activation='relu', return_sequences=True))
aapl_model.add(Dropout(0.4))

aapl_model.add(LSTM(units=120, activation='relu'))
aapl_model.add(Dropout(0.5))

aapl_model.add(Dense(units=1))
#-----------------------------
msft_model = Sequential()

msft_model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_msft_train.shape[1], 1)))
msft_model.add(Dropout(0.2))

msft_model.add(LSTM(units=60, activation='relu', return_sequences=True))
msft_model.add(Dropout(0.3))

msft_model.add(LSTM(units=80, activation='relu', return_sequences=True))
msft_model.add(Dropout(0.4))

msft_model.add(LSTM(units=120, activation='relu'))
msft_model.add(Dropout(0.5))

msft_model.add(Dense(units=1))

#-----------------------------
nflx_model = Sequential()

nflx_model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_nflx_train.shape[1], 1)))
nflx_model.add(Dropout(0.2))

nflx_model.add(LSTM(units=60, activation='relu', return_sequences=True))
nflx_model.add(Dropout(0.3))

nflx_model.add(LSTM(units=80, activation='relu', return_sequences=True))
nflx_model.add(Dropout(0.4))

nflx_model.add(LSTM(units=120, activation='relu'))
nflx_model.add(Dropout(0.5))

nflx_model.add(Dense(units=1))

#-----------------------------
amzn_model = Sequential()

amzn_model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_amzn_train.shape[1], 1)))
amzn_model.add(Dropout(0.2))

amzn_model.add(LSTM(units=60, activation='relu', return_sequences=True))
amzn_model.add(Dropout(0.3))

amzn_model.add(LSTM(units=80, activation='relu', return_sequences=True))
amzn_model.add(Dropout(0.4))

amzn_model.add(LSTM(units=120, activation='relu'))
amzn_model.add(Dropout(0.5))

amzn_model.add(Dense(units=1))

#-----------------------------
meta_model = Sequential()

meta_model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_meta_train.shape[1], 1)))
meta_model.add(Dropout(0.2))

meta_model.add(LSTM(units=60, activation='relu', return_sequences=True))
meta_model.add(Dropout(0.3))

meta_model.add(LSTM(units=80, activation='relu', return_sequences=True))
meta_model.add(Dropout(0.4))

meta_model.add(LSTM(units=120, activation='relu'))
meta_model.add(Dropout(0.5))

meta_model.add(Dense(units=1))

#-----------------------------
nvda_model = Sequential()

nvda_model.add(LSTM(units=50, activation='relu', return_sequences=True, input_shape=(x_nvda_train.shape[1], 1)))
nvda_model.add(Dropout(0.2))

nvda_model.add(LSTM(units=60, activation='relu', return_sequences=True))
nvda_model.add(Dropout(0.3))

nvda_model.add(LSTM(units=80, activation='relu', return_sequences=True))
nvda_model.add(Dropout(0.4))

nvda_model.add(LSTM(units=120, activation='relu'))
nvda_model.add(Dropout(0.5))

nvda_model.add(Dense(units=1))
```

```python
googl_model.compile(loss='mean_squared_error', optimizer='adam')
aapl_model.compile(loss='mean_squared_error', optimizer='adam')
msft_model.compile(loss='mean_squared_error', optimizer='adam')
amzn_model.compile(loss='mean_squared_error', optimizer='adam')
nvda_model.compile(loss='mean_squared_error', optimizer='adam')
meta_model.compile(loss='mean_squared_error', optimizer='adam')
nflx_model.compile(loss='mean_squared_error', optimizer='adam')

googl_model.fit(x_googl_train, y_googl_train, epochs = 50)
aapl_model.fit(x_aapl_train, y_aapl_train, epochs=50)
msft_model.fit(x_msft_train, y_msft_train, epochs=50)
amzn_model.fit(x_amzn_train, y_amzn_train, epochs=50)
nvda_model.fit(x_nvda_train, y_nvda_train, epochs=50)
meta_model.fit(x_meta_train, y_meta_train, epochs=50)
nflx_model.fit(x_nflx_train, y_nflx_train, epochs=50)
```
